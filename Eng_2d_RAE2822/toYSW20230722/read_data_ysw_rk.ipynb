```python
import numpy as np
import matplotlib.pyplot as plt

data_tot=np.load('RAE2822_c81table_qoi_ysw20230722.npy')
Ma_grid=np.load('RAE2822_c81table_ma_grid_ysw20230722.npy')
AoA_grid=np.load('RAE2822_c81table_aoa_grid_ysw20230722.npy')
```

```python
label_list=['LF', 'MF', 'HF']
linestyle_list=['-', '--', '-.']
plt.rcParams['font.family']='times new roman'
plt.rcParams['font.size']=15
for qoi in range(3):
    plt.figure(figsize=(15,13))
    plt.subplots_adjust(hspace=0.4, wspace=0.4)
    for i in range(9):
        plt.subplot(3,3,i+1)
        for j in range(3):
            plt.plot(AoA_grid, data_tot[j,qoi,i,:], label=label_list[j], linestyle=linestyle_list[j], alpha=0.8)
        plt.legend(frameon=False)
        plt.title('Ma: {:.2f}'.format(Ma_grid[i]))
        plt.xlabel('Angle of attack')
    plt.show()
```

```python
title_list=['$C_l$', '$C_d$', '$C_m$']
plt.figure(figsize=(15,4))
plt.subplots_adjust(hspace=0.4, wspace=0.4)
for qoi in range(3):
    plt.subplot(1,3,qoi+1)
    for j in range(3):
        plt.contour(AoA_grid, Ma_grid, data_tot[j,qoi], linestyles=linestyle_list[j], alpha=0.8)
    plt.xlabel('Angle of attack')
    plt.ylabel('Mach number')
    plt.title(title_list[qoi])
plt.show()
```
